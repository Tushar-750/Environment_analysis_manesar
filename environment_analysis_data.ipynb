```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
random.seed(42)
```

```python
## üå¶Ô∏è Simulated Weather Data
weather_data = {
    "temperature": round(random.uniform(24, 40), 1),
    "humidity": random.randint(40, 90),
    "description": random.choice(["Sunny", "Cloudy", "Rainy", "Hazy"])
}
print("üå¶Ô∏è Weather Report")
print(f"Temperature: {weather_data['temperature']} ¬∞C")
print(f"Humidity: {weather_data['humidity']} %")
print(f"Condition: {weather_data['description']}")
```

```python
## üí® Simulated Air Quality Data
air_quality_data = {
    "aqi": random.randint(50, 250),
    "main_pollutant": random.choice(["PM2.5", "PM10", "O3", "NO2"])
}
print("\nüí® Air Quality Report")
print(f"AQI: {air_quality_data['aqi']}")
print(f"Main Pollutant: {air_quality_data['main_pollutant']}")
```

```python
## land use distribution in manesar 
import pandas as pd
import random
import matplotlib.pyplot as plt
land_use = pd.DataFrame({
    "Type": ["Agriculture", "Urban", "Forest", "Water Bodies"],
    "Percent": [random.randint(30, 45), random.randint(25, 40), random.randint(10, 25), random.randint(5, 10)]
})
land_use["Percent"] = (land_use["Percent"] / land_use["Percent"].sum()) * 100
print(land_use)
land_use.plot.pie(
    y="Percent",
    labels=land_use["Type"],
    autopct="%1.1f%%",
    startangle=140,
    legend=False
)
plt.title("Land Use Distribution in Manesar (Simulated)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
## Socio_economic indicators in manesar
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
socio_data = pd.DataFrame({
    "Parameter": ["Literacy Rate", "Population (lakhs)", "Employment Rate", "Urbanization Rate"],
    "Value": [
        random.randint(75, 90),
        round(random.uniform(0.8, 1.5), 1),
        random.randint(65, 85),
        random.randint(50, 80)
    ]
})
print(socio_data)
sns.barplot(data=socio_data, x="Parameter", y="Value", hue="Parameter", palette="coolwarm", legend=False)
plt.title("Socio-Economic Indicators in Manesar (Simulated)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
